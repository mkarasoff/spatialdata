# This is a notebook that will fix paths in a mosaic data set
It provides some help in finding out what the wrong paths are to begin with and then outputs the fixed paths when done.

These environment vairables need to be set properly

`projDir` is the path to the project.

`mdsPath` is the full path to the MDS needing the fix.

`badPath` is the path to fix.

`goodPath` is what to set the new path to.


```python
import arcpy
import os

projDir=r'c:\Users\mike\Documents\ArcGIS\Projects\Assignements\remoteSensing\week4'

mdsPath = projDir + r'\data\SampleData\Vexcel-Hollywood\0_MosaicDataset\VexcelHollywood.gdb\VexcelHollywood_Stereo'

badPath = r'c:\SampleData'
goodPath = projDir + r'\data\SampleData'

mdsName = os.path.basename(mdsPath)
pathTblName = "%s_pathTable.dbf"  % ( mdsName )
dataPathTbl="%s\\%s"  % ( projDir, pathTblName )
#dataPathTbl=projDir + r'\peerGrade\peerGradeW4\Default.gdb\VexcelHollywood_Stereo_Paths'
```

A function to list the paths.

```python
def listPaths():
    print(mdsPath)
    print(dataPathTbl)
    arcpy.management.ExportMosaicDatasetPaths(mdsPath, dataPathTbl, '', "ALL", "RASTER;ITEM_CACHE")
    fieldnames = [f.name for f in arcpy.ListFields(dataPathTbl)]
    cursor = arcpy.da.SearchCursor(dataPathTbl, fieldnames)
    for row in cursor:
        print(row)
        
    arcpy.RemoveLayer(pathTblName)
    os.remove(dataPathTbl)

    return
```

List paths to get an idea what is going on.

```python
listPaths()
```

This will repair the paths

```python
arcpy.management.RepairMosaicDatasetPaths(mdsPath, badPath + " " + goodPath, '')

listPaths()
```
